```python
def fensu(abc):  #对每日数据Series进行MAD值整理、Z-Score标准、归一化打分。越低分数越高！
    X = abc
    X_mean = np.median(X, axis=0)       #求中位数
    # calculate variance  
    X_mad = np.median(abs(X - X_mean))  #偏差取绝对值后再求中位数，即中位绝对偏差MAD
    X_std = X.std(axis=0)               #求标准差
    X_max = X_mean + 3 * X_mad          #求上限
    X_min = X_mean - 3 * X_mad          #求下限。
    X_return = (np.clip(X, X_min, X_max) - X_mean) / X_std  #去极值处理，再标准化
    r_max = max(X_return)
    r_min = min(X_return)
    x_fensu = (r_max - X_return) / (r_max - r_min)#归一化。此处是：值越低，分数越高！
    return (x_fensu)
pricedf_100 = fensu(X)#对价格矩阵进行打分。
pricedf_100
```
