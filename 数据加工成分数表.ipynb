```python
import pandas as pd
datename = '180101To220825'
pricefilename = '{}{}{}'.format('data/price_',datename,'.pkl')
premfilename = '{}{}{}'.format('data/prem_',datename,'.pkl')
sizefilename = '{}{}{}'.format('data/size_',datename,'.pkl')
price100filename = '{}{}{}'.format('data/price100_',datename,'.pkl')
prem100filename = '{}{}{}'.format('data/prem100_',datename,'.pkl')
size100filename = '{}{}{}'.format('data/size100_',datename,'.pkl')

pricedf=pd.read_pickle(pricefilename) #直接读取pkl数据文件。
pricedf
```

```python
import numpy as np
def fensu01(abc):#数值小，分数大！
    dt = abc.T
    median = dt.quantile(0.5)   # 找出中位数
    new_median = (abs((dt - median)).quantile(0.5))   # 偏差值的中位数
    dt_up = median + 3*new_median    # 上限
    dt_down = median - 3*new_median  # 下限
    X_return = dt.clip(dt_down, dt_up, axis=1)    # 超出上下限的值，赋值为上下限
    r_max = np.max(X_return)
    r_min = np.min(X_return)
    X_fensu01 = (r_max-X_return)/(r_max-r_min)
    Y = X_fensu01.T
    return Y
```

```python
price100df = fensu01(pricedf)
price100df
```

```python
premdf=pd.read_pickle(premfilename)
premdf
```

```python
prem100df = fensu01(premdf)
prem100df
```

```python
sizedf=pd.read_pickle(sizefilename)
sizedf
```

```python
size100df = fensu01(sizedf)
size100df
```

```python
def df2pkl():   #数据加工后的输出函数。
    price100df.to_pickle(price100filename)
    prem100df.to_pickle(prem100filename)
    size100df.to_pickle(size100filename)
    return
#df2pkl()    #取消函数名前的"#"则将数据输出至pkl文件。
```
