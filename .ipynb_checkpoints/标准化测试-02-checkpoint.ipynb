```python
pricedf_tf = pricedf_t.fillna(pricedf_t.median())#缺失值中值填充（按日期）
pricedf_tf
```

```python
# MAD法去极值
def extreme_MAD(dt,n):
    median = dt.quantile(0.5)   # 找出中位数
    new_median = (abs((dt - median)).quantile(0.5))   # 偏差值的中位数
    dt_up = median + n*new_median    # 上限
    dt_down = median - n*new_median  # 下限
    return dt.clip(dt_down, dt_up, axis=1)    # 超出上下限的值，赋值为上下限

pricedf_t_01mad = extreme_MAD(pricedf_t,3)
pricedf_t_01mad.head(7)
```

```python
pricedf_t_01mad['2021-12-16'].describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40, 5))
fund0 = pricedf_t['2021-12-16']
fund = pricedf_t_01mad['2021-12-16']
plt.plot(fund.index,fund0, color="red")
plt.plot(fund.index,fund)

plt.title("pe_ratio vs pe_ratio_fixed")
plt.show()
```

```python
# Z值标准化
def standardize_z(dt):
    median = dt.median()    #  截面数据均值
    std = dt.std()       #  截面数据标准差
    return (dt - median)/std
pricedf_t_02STD=standardize_z(pricedf_t_01mad)
pricedf_t_02STD.head(7)
```
